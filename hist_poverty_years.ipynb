```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
merged_personal_df = pd.read_csv("./Resources/merged_personal_df.csv")
```

```python
merged_personal_df
```

```python
mp2014_df = merged_personal_df.loc[(merged_personal_df["Year"] == 2014)]
mp2015_df = merged_personal_df.loc[(merged_personal_df["Year"] == 2015)]
```

```python
# convert crime count to crime rate
# crime rate vs home value
# crime rate vs poverty rate
# per year
# 10 graphs
```

```python
population1 = mp2014_df["PC"]
population2 = mp2015_df["PC"]
plt.subplot(2, 1, 2)
plt.hist(population1, 49, density=True, alpha=0.7, label="population1")
plt.hist(population2, 49, density=True, alpha=0.7, label="population2")
plt.axvline(population1.mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(population2.mean(), color='k', linestyle='dashed', linewidth=1)
plt.legend()  
```
